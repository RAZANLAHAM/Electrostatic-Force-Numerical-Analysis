```python
# Electrostatic-Force-Numerical-Analysis
import math
import numpy as np
import matplotlib.pyplot as plt

# Physical constants and input values
q = 2e-5
Q = 2e-5
a = 0.9
F_target = 1.0
eps0 = 8.85e-12

# Electrostatic force function
def F(x):
    return (1 / (4 * math.pi * eps0)) * (q * Q * x) / ((a**2 + x**2)**(3/2))

# Root function: f(x) = F(x) - F_target
def f(x):
    return F(x) - F_target

# Derivative of f(x) for Newton-Raphson
def df(x):
    numerator = q * Q * ((a**2 + x**2)**(3/2)) - q * Q * x * 3 * x * (a**2 + x**2)**(1/2)
    denominator = (a**2 + x**2)**3
    return (1 / (4 * math.pi * eps0)) * (numerator / denominator)

# Bisection method with iteration logging
def bisection_method_verbose(x_low, x_high, tol=1e-6, max_iter=100):
    iterations = []
    for i in range(max_iter):
        x_mid = (x_low + x_high) / 2
        f_mid = f(x_mid)
        error = abs(x_high - x_low) / 2
        iterations.append((i + 1, x_mid, f_mid, error))

        if abs(f_mid) < tol or error < tol:
            break

        if f(x_low) * f_mid < 0:
            x_high = x_mid
        else:
            x_low = x_mid

    print("Bisection Method:")
    print("Iter\t x\t\t f(x)\t\t Error")
    for (i, x_val, f_val, err) in iterations:
        print(f"{i}\t {x_val:.6f}\t {f_val:.3e}\t {err:.3e}")

    print(f"\nFinal x = {x_mid:.6f}, iterations = {i+1}, final error ≈ {error:.2e}")
    return x_mid, i + 1

# Newton-Raphson method with iteration logging
def newton_raphson_verbose(x0, tol=1e-6, max_iter=100):
    x = x0
    iterations = []

    for i in range(max_iter):
        fx = f(x)
        dfx = df(x)
        if dfx == 0:
            raise ZeroDivisionError("Zero derivative encountered.")
        x_new = x - fx / dfx
        error = abs(x_new - x)
        iterations.append((i + 1, x_new, f(x_new), error))
        if abs(f(x_new)) < tol or error < tol:
            break
        x = x_new

    print("\nNewton-Raphson Method:")
    print("Iter\t x\t\t f(x)\t\t Error")
    for (i, x_val, f_val, err) in iterations:
        print(f"{i}\t {x_val:.6f}\t {f_val:.3e}\t {err:.3e}")

    print(f"\nFinal x = {x_new:.6f}, iterations = {i+1}, final error ≈ {error:.2e}")
    return x_new, i + 1

# Run both methods
x_bisect, it_bisect = bisection_method_verbose(0.01, 2)
x_newton, it_newton = newton_raphson_verbose(1.0)

# Plotting the force function and roots
x_vals = np.linspace(0.01, 2, 300)
f_vals = [F(x) for x in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='F(x)', linewidth=2)
plt.axhline(y=F_target, color='red', linestyle='--', label='Target Force')
plt.axvline(x=x_bisect, color='green', linestyle='--', label='Bisection Result')
plt.axvline(x=x_newton, color='orange', linestyle='--', label='Newton-Raphson Result')
plt.xlabel("x (m)")
plt.ylabel("F(x) (N)")
plt.title("Electrostatic Force vs Distance")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
